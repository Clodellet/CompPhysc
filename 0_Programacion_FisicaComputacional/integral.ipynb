```python
import numpy as np
import matplotlib.pyplot as plt
from math import *

%matplotlib notebook
```

```python
def f(x):
    return -10*x**2 -7*x + 500

def F(x):
    return -(10/3)*x**3 - (7/2)*x**2 + 500*x
```

```python
def g(x):
    return sin(x)**2+x+3

def G(x):
    return (x**2)/2 + (7/2)*x -(sin(2*x)/4)
```

```python
def h(x):
    return exp(2*x)

def H(x):
    return exp(2*x)/2
```

```python
def j(x):
    return 4*x**3
    
def J(x):
    return x**4
```

```python
a = -7.5
b = 7.5
N = 100000

dx =(b-a)/N 
```

```python
S = (f(a) + f(b))/2

for k in range(1, N):
    S = S + f(a+ (k*dx))

I = S*dx
print(I)
```

```python
print(F(b)-F(a))
```

```python
S = (g(a) + g(b))/2

for k in range(1, N):
    S = S + g(a+ (k*dx))

I = S*dx
print(I)
print (G(b)-G(a))
```

```python
S = (h(a) + h(b))/2

for k in range(1, N):
    S = S + h(a+ (k*dx))

I = S*dx
print(I)
print (H(b)-H(a))
```

```python
S = (j(a) + j(b))/2

for k in range(1, N):
    S = S + j(a+ (k*dx))

I = S*dx
print(I)
print (J(b)-J(a))
```
